```python
features=list(train_df.columns)
features.remove('Survived')
features.remove('Name')
X=train_df[features]
X['Age']=X['Age'].fillna(0)
y=train_df['Survived']
print(X.shape)
print(y.shape)
```

```python
sex_df=pd.get_dummies(X['Sex'], drop_first=True)
embarked_df=pd.get_dummies(X['Embarked'])
X=pd.concat([X, sex_df, embarked_df], axis=1)
X=X.drop(columns=['Embarked', 'Sex'])
X.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test=train_test_split(X, y)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
scaler.fit(X_train)
X_train_scaled=scaler.transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf=RandomForestClassifier()
rf.fit(X_train_scaled, y_train)
print(f'Train Score: {rf.score(X_train_scaled, y_train)}')
print(f'Test Score: {rf.score(X_test_scaled, y_test)}')
```
